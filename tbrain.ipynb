```python
X_train,X_test,y_train,y_test = train_test_split(X, y,test_size=0.25, random_state=42)
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=42)
# from sklearn.ensemble import GradientBoostingClassifier #梯度提升 會掛掉
# clf = GradientBoostingClassifier(random_state=42)
# from sklearn.ensemble import RandomForestClassifier #隨機森林 也會掛掉
# clf = RandomForestClassifier(n_estimators=100, n_jobs = -1, random_state = 42, min_samples_leaf = 10)

clf.fit(X_train,y_train)
```

```python
predictions = clf.predict_proba(X_train.drop_duplicates(subset=['chid']))
```
