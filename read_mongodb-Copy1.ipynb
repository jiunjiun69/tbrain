```python
import pymongo
from pymongo import MongoClient
import numpy as np
import pandas as pd
```

```python
#顯示所有列
pd.set_option('display.max_columns', None)
```

```python
MONGO_HOST = 'ip'
MONGO_DB = 'jiunjiun'
MONGO_COLLETION = 'tbrain_cc_training_48tags_hash_final'
```

```python
def connect_mongo():  #連線資料庫
    global collection
    client = MongoClient(MONGO_HOST, 27017, username = '', password = '')
    db = client[MONGO_DB]
    collection = db[MONGO_COLLETION]
```

```python
connect_mongo()  #呼叫連線資料庫函式
query = { 'chid' : '10128239' }
cursor = collection.find(query)  #依query查詢資料
stock =  pd.DataFrame(list(cursor))  #轉換成DataFrame
del stock['_id']  #刪除欄位_id
```
