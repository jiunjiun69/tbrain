```python
import pymongo
import csv
import time
def connect_mongo():
    client = pymongo.MongoClient('ip', 27017, username = '', password = '')
    db = client.jiunjiun
    collection = db['tbrain_cc_training_48tags_hash_final']
    return collection

def insertToMongoDB(set1):
    with open('tbrain_cc_training_48tags_hash_final.csv','r',encoding='utf-8') as csvfile:
        # 調用csv中的DictReader函數直接獲取資料為字典形式
        reader = csv.DictReader(csvfile)
        csv_data = []
        # 創建一個counts計數一下 看自己一共添加了了多少條資料
        counts = 0
        index = 1
        for each in reader:
            csv_data.append(each)
            if index==10000:#10000個之後寫入MongoDB中
                set1.insert(csv_data)
                csv_data.clear()
                index = 0
                print("成功添加了" + str(counts) + "條數據")
            counts+=1
            index+=1
        if len(csv_data)>0:#剩餘的數據
            set1.insert(csv_data)
            print("成功添加了%s條數據"%len(csv_data))

if __name__=='__main__':
    print(time.strftime('%Y-%m-%d %H:%M:%S'))#計算時間用
    set1 = connect_mongo()
    insertToMongoDB(set1)
    print(time.strftime('%Y-%m-%d %H:%M:%S'))
```
