```python
DTC.fit( X_Train[SelectedFeatures], Y_Train )
```

```python
predictions = DTC.predict_proba( X_Train[SelectedFeatures].drop_duplicates(subset=['chid']) )
```

```python
best_predictions = np.argsort(predictions, axis=1)[:,-49:]
```

```python
best_predictions[0]
```

```python
best_predictions_final = np.zeros((len(best_predictions), 3), dtype=int)
```

```python
for a in range(len(best_predictions)):
    sum = 0
    for i in range(49): 
        if (best_predictions[a, i] == [2,6,10,12,13,15,18,19,21,22,25,26,39,37,39,48]).any():
            best_predictions_final[a, sum] = int(best_predictions[a, i])
            sum += 1 
        if sum == 3: 
            break
```
