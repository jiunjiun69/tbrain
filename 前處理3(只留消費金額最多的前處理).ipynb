```python
# 讀前處理過後的資料
train = pd.read_csv('Processed_tbrain_training.csv')
print(train.shape)
print(train.info())
train.head(5)
```

```python
# 把每位顧客每個月消費金額最多的資料留下，不是消費金額最多的不留
train = train.iloc[train.groupby(['chid','dt']).apply(lambda x: x['txn_amt'].idxmax())]
```
