```python
import numpy as np
import pandas as pd
```

```python
#顯示所有列
pd.set_option('display.max_columns', None)
```

```python
# 讀前處理過後的資料
train = pd.read_csv('Processed_tbrain_training.csv')
print(train.shape)
print(train.info())
train.head(5)
```

```python
# 對卡片消費次數做加總，並刪掉個別消費次數，這邊可能要再試
train["sum_card_txn_cnt"] = train["card_1_txn_cnt"] + train["card_2_txn_cnt"] + train["card_3_txn_cnt"] + train["card_4_txn_cnt"] + train["card_5_txn_cnt"] + train["card_6_txn_cnt"] + train["card_7_txn_cnt"] + train["card_8_txn_cnt"] + train["card_9_txn_cnt"] + train["card_10_txn_cnt"] + train["card_11_txn_cnt"] + train["card_12_txn_cnt"] + train["card_13_txn_cnt"] + train["card_14_txn_cnt"] + train["card_other_txn_cnt"]
train = train.drop(columns = ['card_1_txn_cnt','card_2_txn_cnt','card_3_txn_cnt','card_4_txn_cnt','card_5_txn_cnt','card_6_txn_cnt','card_7_txn_cnt','card_8_txn_cnt','card_9_txn_cnt','card_10_txn_cnt','card_11_txn_cnt','card_12_txn_cnt','card_13_txn_cnt','card_14_txn_cnt','card_other_txn_cnt'])
```

```python
# 對卡國內消費次數做加總，並刪掉個別國內消費次數
train["sum_domestic_cnt"] = train["domestic_offline_cnt"] + train["domestic_online_cnt"]
train = train.drop(columns = ['domestic_offline_cnt','domestic_online_cnt'])
```

```python
# 對海外消費次數做加總，並刪掉個別海外消費次數
train["sum_overseas_cnt"] = train["overseas_offline_cnt"] + train["overseas_online_cnt"]
train = train.drop(columns = ['overseas_offline_cnt','overseas_online_cnt'])
```

```python
print(train.shape)
print(train.info())
train.head(5)
```

```python
# 刪掉txn_cnt消費次數
train = train.drop(columns = ['txn_cnt'])
```

```python
print(train.shape)
print(train.info())
train.head(5)
```

```python
# 刪完加總後的消費次數資料後再存檔一次
filename = 'Processed_tbrain_training_sum_cnt.csv'
train.to_csv(filename,index=False)
```
